```python
import pandas as pd
```

```python
# 거래 데이터 불러오기
df = pd.read_csv("transaction.csv")

print(df.shape)
df.head()
```

```python
# 데이터 1000개 추출
df = df.head(1000)

print(df.shape)
df.head()
```

groupby

```python
df.groupby('id').sum()
```

```python
# index를 칼럼으로 지정
df.groupby('id').sum().reset_index()
```

```python
df.groupby('id').describe()
```

Index

```python
# id 칼럼을 인덱스로 지정
df1 = df.set_index('id')

df1.head()
```

```python
# 인덱스 버리기
df1.reset_index(drop=True).head()
```

```python
# index로 지정된 id 칼럼을 칼럼으로 바꾸기
df1.reset_index().head()
```

```python
# index였던 id 칼럼이 사라지고 0부터 시작되는 새로운 인덱스 생성
df1.reset_index(drop=True)
```
