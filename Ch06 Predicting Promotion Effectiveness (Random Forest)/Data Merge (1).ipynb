```python
import pandas as pd
```

```python
pd.DataFrame({'key1' : [1, 2, 3, 4, 5]})
```

```python
# 데이터 프레임 생성
df1 = pd.DataFrame({'A' : [1, 3, 5, 7, 9]
                    , 'B' : [11, 12, 13, 14, 15]
                    , 'C' : [0, 1, 2, 3, 4]
                    , 'D' : [101, 102, 103, 104, 105]}
                    , index=[1, 2, 3, 4, 5])

df1
```

```python
df2 = pd.DataFrame({'A' : [2, 4, 6, 8, 10]
                    , 'B' : [1, 2, 3, 4, 5]
                    , 'C' : [1, 2, 3, 4, 5]
                    , 'D' : [1011, 1021, 1031, 1041, 1051]}
                    , index=[6, 7, 8, 9, 10])

df2
```

# 1. concat

```python
# 주로 행(axis=0)을 기준으로 데이터프레임을 합치는 데 쓰임
# 리스트 형태로 데이터 적재
pd.concat([df1, df2])
```

```python
# concat을 column 기준으로 붙였을 때는 index가 같은 것끼리 묶임
pd.concat([df1, df2], axis=1)
```

```python
# 인덱스 조정 후 concat
df1.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)

pd.concat([df1, df2], axis=1)
```
