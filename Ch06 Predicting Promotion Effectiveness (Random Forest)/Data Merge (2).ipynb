```python
import pandas as pd
```

```python
# 데이터 프레임 생성
df1 = pd.DataFrame({'id' : ['x01', 'x02', 'x03', 'x04', 'x05']
                    , 'A' : [1, 3, 5, 7, 9]
                    , 'B' : [11, 12, 13, 14, 15]
                    , 'C' : [0, 1, 2, 3, 4]
                    , 'D' : [101, 102, 103, 104, 105]})

df1
```

```python
df2 = pd.DataFrame({'id' : ['x01', 'x02', 'x03', 'x04', 'x06']
                    , 'E' : [1, 3, 5, 7, 9]
                    , 'F' : [11, 12, 13, 14, 15]
                    , 'G' : [0, 1, 2, 3, 4]
                    , 'H' : [101, 102, 103, 104, 105]})

df2
```

# 2. merge

```python
# 1) inner join (디폴트 값)
pd.merge(df1, df2, on='id')
```

```python
pd.merge(df1, df2, on='id', how='inner')
```

```python
# 2) outer join
pd.merge(df1, df2, on='id', how='outer')
```

```python
# 3) left join
pd.merge(df1, df2, on='id', how='left')
```

```python
# 4) right join
pd.merge(df1, df2, on='id', how='right')
```

primary키가 다른 경우

```python
df3 = pd.DataFrame({'user_id' : ['x01', 'x02', 'x03', 'x04', 'x06']
                    , 'E' : [1, 3, 5, 7, 9]
                    , 'F' : [11, 12, 13, 14, 15]
                    , 'G' : [0, 1, 2, 3, 4]
                    , 'H' : [101, 102, 103, 104, 105]})

df3
```

```python
pd.merge(df1, df3, left_on='id', right_on='user_id')
```

# 3. join

```python
# 데이터 프레임 생성
df1 = pd.DataFrame({'A' : [1, 3, 5, 7, 9]
                    , 'B' : [11, 12, 13, 14, 15]
                    , 'C' : [0, 1, 2, 3, 4]
                    , 'D' : [101, 102, 103, 104, 105]}
                    , index=[1, 2, 3, 4, 5])

df1
```

```python
df2 = pd.DataFrame({'E' : [2, 4, 6, 8, 10]
                    , 'F' : [1, 2, 3, 4, 5]
                    , 'G' : [1, 2, 3, 4, 5]
                    , 'H' : [1011, 1021, 1031, 1041, 1051]}
                    , index=[6, 7, 8, 9, 10])

df2
```

```python
# left join이 디폴트
# key값이 따로 정해져있지 않고 index를 기준으로 join
df1.join(df2)
```

```python
# # 인덱스 조정 후 join
df1.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)

df1.join(df2)
```

```python
# right join
df1.join(df2, how='right')
```
